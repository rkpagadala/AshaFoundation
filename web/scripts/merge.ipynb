```python
df = pd.read_csv("../data/per_population_state_year.csv")


df['chapter'] = 'All Chapters'
df
```

```python
population_df = pd.read_csv("../data/population.csv")
```

```python
df_list = []  # Create an empty list to collect rows as dictionaries

for i, row in per_pop_df.iterrows():
    # Calculate total_amount
    total_amount = per_pop_df[(per_pop_df['year'] == row['year']) & 
                              (per_pop_df['chapter'] == row['chapter']) &
                              (per_pop_df['state'] != "BIMARU")
                              ]['amount'].sum()
    funds_percent = (row['amount'] / total_amount) * 100
    if row['state'] == "BIMARU":
        population_percent = 36.779
    else:
        population_percent = population_df[population_df['state'] == row['state']]['% of Total'].values[0]
    # Append row data as a dictionary to df_list
    df_list.append({
        "state": row['state'],
        "year": row['year'],
        "state_total_amount": row['amount'],
        "total_amount": total_amount,
        "state_percentage": funds_percent,
        "chapter": row['chapter'],
        "pop_adj_units" : (funds_percent / population_percent)
    })

# Convert the list of dictionaries to a DataFrame
chapter_wise_df = pd.DataFrame(df_list)
chapter_wise_df
```

```python

# Display df
df = pd.concat([df, chapter_wise_df])
df
```
